```python
data_beforeC=data.copy()
data_beforeC.head()
```

```python
def func (x) :
    if (x==15):
        return -1
    else :
        return x
def func_day (x):
    if(x==400):
        return 0
    else :
        return x
```

```python
data.loc[:,'year']=data['year'].apply(func)
data.loc[:,'day']=data['day'].apply(func_day)
```

* We calculate the correlations related to day and year 

```python
correlation_dayyear_0=pg.pairwise_corr(data_beforeC.drop('id',axis=1), method='pearson')
feautures=['day','year']
correlation_dayyear_0=correlation_given_feautures(correlation_dayyear_0, feautures)
```

```python
correlation_dayyear_1=pg.pairwise_corr(data.drop('id',axis=1), method='pearson')
feautures=['day','year']
correlation_dayyear_1=correlation_given_feautures(correlation_dayyear_1, feautures)
```

* I create a function to find the difference between correlation coefficients.

```python
def diff_of_correlation(correlation_0,correlation_1):
    correlation_0.sort_values(['X','Y'],ascending=False,inplace=True)
    correlation_0=correlation_0.reset_index(drop=True)
    correlation_1.sort_values(['X','Y'],ascending=False,inplace=True)
    correlation_1=correlation_1.reset_index(drop=True)
    
    correlation_diff=correlation_0.loc[:,['X','Y','r']]
    correlation_diff.loc[:,'r_1']=correlation_1.loc[:,'r']
    corr_diff=(correlation_1.loc[:,'r']-correlation_0.loc[:,'r'])
    correlation_diff.loc[:,'diff']=corr_diff
    
    correlation_diff.loc[:,'abs_diff']=correlation_diff.loc[:,'diff'].apply(abs)
    correlation_diff.sort_values('abs_diff',ascending=False,inplace=True)
    correlation_diff=correlation_diff.reset_index(drop=True)
    return correlation_diff
```

```python
diff_corr=diff_of_correlation(correlation_dayyear_0,correlation_dayyear_1)

diff_corr
```

* We can see that feautures that has changed most are categorized features. since day and year for freezer rats has changed dramatically they affected most to the correlations with the other categorized features. Kmethod and place correlations with day and year are the most changed ones and they all have changed their sign of correlation. Because also year and day for freezer rats has 'changed' sign it seems intuitive.
